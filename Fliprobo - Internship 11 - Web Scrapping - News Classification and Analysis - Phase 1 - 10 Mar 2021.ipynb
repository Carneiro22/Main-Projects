# Jan-01-2018

```python
# Connecting to the web driver
driver=webdriver.Chrome(r"C:\Users\Acer\Desktop\Internship 11 - Project 1 - Avinash Carneiro\chromedriver.exe")

# Specifying the url of the webpage
driver.get("https://timesofindia.indiatimes.com/2018/1/1/archivelist/year-2018,month-1,starttime-43101.cms")
```

```python
import requests
 
url = "https://timesofindia.indiatimes.com/2018/1/1/archivelist/year-2018,month-1,starttime-43101.cms"
 
# Getting the webpage, creating a Response object.
response = requests.get(url)
 
# Extracting the source code of the page.
data = response.text
 
# Passing the source code to BeautifulSoup to create a BeautifulSoup object for it.
soup = BeautifulSoup(data, 'html')
 
# Extracting all the <a> tags into a list.
tags = soup.find_all('a')
 
# Extracting URLs from the attribute href in the <a> tags.
Links=[]
for tag in tags:
    Links.append(tag.get('href'))
```

```python
# Article links in the month of Jan 2018

Jan_01=Links[316:815]
Jan_01
```

```python
# Creating empty lists
Date=[]
Author=[]
Vertical=[]
Headline=[]
Description=[]
```

```python
#loop for all the required data of product
for url in Jan_01:
    driver.get(url)

#To find the Date of the Article
    try: 
        date=driver.find_element_by_xpath("//div[@data-plugin='dateformat']")
        Date.append(date.text.replace('\r\n','').replace('Created:','').replace('Updated:',''))
    except NoSuchElementException   as e:
        Date        
    try:
        date=driver.find_element_by_xpath("//div[@class='_3Mkg- byline']")
        Date.append(date.text.replace('\r\n','').replace('Created:','').replace('\n',''))
    except NoSuchElementException   as e:
        Date.append("-")
        
        
#To find the Author of the Article 
    try: 
        author=driver.find_element_by_xpath("//a[@class='auth_detail']")
        Author.append(author.text.replace('\r\n','').replace('Created:','').replace('\n',''))
    except NoSuchElementException   as e: 
        Author 
    try:    
        author=driver.find_element_by_xpath("//div[@class='_3Mkg- byline']")
        Author.append(author.text.replace('\r\n','').replace('Created:','').replace('\n',''))
        
    except NoSuchElementException   as e:
        Author.append("-")
        
    
#To find the Headline of the Article  
    try: 
        headline=driver.find_element_by_xpath("//h1[@data-plugin='arttitle']")
        Headline.append(headline.text.replace('\r\n','').replace('Created:','').replace('\n',''))
    except NoSuchElementException   as e: 
        Headline   
    try: 
        headline=driver.find_element_by_xpath("//h1[@class='_23498']")
        Headline.append(headline.text.replace('\r\n','').replace('Created:','').replace('\n',''))
    except NoSuchElementException   as e:
        Headline.append("-")
        

#To find the Vertical of the Article
    try: 
        vertical=driver.find_element_by_xpath("//p[@class='_3O-OX']")
        Vertical.append(vertical.text.replace('\r\n','').replace('»',' ').replace('\n',''))    
    except NoSuchElementException   as e:
        Vertical.append("-")
        
                        
#To find the Description of the Article
    try: 
        description=driver.find_element_by_xpath("//div[@class='Normal']")      
        Description.append(description.text.replace('\r\n','').replace('Created:','').replace('\n','')) 
    except NoSuchElementException   as e: 
        Description                   
    try:
        description=driver.find_element_by_xpath("//div[@class='ga-headlines']")
        Description.append(description.text.replace('\r\n','').replace('Created:','').replace('\n',''))
    except NoSuchElementException   as e:
        Description.append("-")
```

```python
#Creating dataframe for scraped data

df1=pd.DataFrame({})
df2=pd.DataFrame({})
df3=pd.DataFrame({})
df4=pd.DataFrame({})
df5=pd.DataFrame({})

df1['Author']=Author
df1['Author']=df1['Author'].apply(lambda x: x.replace('|','').replace('Updated:','').replace('toi ','')
                                  .replace('Jan','').replace(',','').replace('IST','').replace(':','').replace('-',''))
df1['Author']=df1['Author'].str.replace('\d+', '')

df2['Date']=Date
df2['Date']=df2['Date'].apply(lambda x: x.replace('|','').replace('&','').replace('Updated:','').replace('TNN ','').replace('toi ','').replace('Jan','01,').replace('Dec','12,'))
df2['Date'] = df2['Date'].str.replace(r"[a-zA-Z]",'')

df3['Vertical']=Vertical
df3['Vertical'].str.replace('\d+', '')

df4['Headline']=Headline
df4['Headline'].str.replace('\d+', '')
df5['Description']=Description

#Creating dataframe for Jan_01_2018
Jan_01_2018 = pd.concat([df1, df2, df3, df4, df5], axis=1)
```
